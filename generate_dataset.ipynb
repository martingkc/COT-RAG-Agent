````python
async def get_qa_ds(questions: list):
    """Gathers requests and executes them asyncly returns map where key = q and val = a"""

    tasks = [get_completion(q) for q in questions]

    res = await asyncio.gather(*tasks)
    docs = []
    for q, a in zip(questions, res):
        docs.append({"question": q, "answer": f"""{a}"""})
    
    return docs
````
